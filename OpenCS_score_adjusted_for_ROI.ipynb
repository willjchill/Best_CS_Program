### OpenCS.com uses a few metrics to produced a normalized score to rank each CS university in the US

I want to include ROI rankings to better improve these scores overall. We will use python to extract the data and analyze it.

```python
import requests
import re
```

```python
def get_html_content(url):
    """
    Fetches the HTML content of a website.

    Args:
        url (str): The URL of the website.

    Returns:
        str: The HTML content of the website, or None if an error occurs.
    """
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL '{url}': {e}")
        return None

# Example usage:
website_url = "https://www.payscale.com/college-roi/page/"
html_content = ""
for i in range(1, 188):
  html_content += get_html_content(website_url + f"{i}")

# if html_content:
#     # Process the HTML content (e.g., parse it with BeautifulSoup)
#     print(html_content)
```

```python
# A function to remove all tags from a string
def remove_tags(data_element):
  fixed_data_element = ""
  for i in range(len(data_element)):
    if(data_element[i] == "<"):
      index_end_bracket = data_element.find(">", i)
      i = index_end_bracket
      fixed_data_element = data_element[i+1:]
  return fixed_data_element
```

```python
# Getting a list of all the universities from ROI website

university_tag = '<span class="roi-grid__schoolname--text">'
ending_tag = '</a>'
positions = [match.start() for match in re.finditer(university_tag, html_content)]
print(len(positions))
first_university = html_content[positions[0]:html_content.find(ending_tag, positions[0])] # DEBUG
```

```python
print(remove_tags(first_university)) # DEBUG
```

```python
# Now that we know the simple case works, we want to generate a list of POSITIONS for every page
list_of_universities = []
for i in range(len(positions)):
  list_of_universities.append(remove_tags(html_content[positions[i]:html_content.find(ending_tag, positions[i])]))
print(list_of_universities)
```

List of Universities variable contains a list of universities. Their index is the score with respect to ROI. We will factor that into the OpenCS ranking now.

```python
def remove_tags_alt(data_element):
  if(data_element.find('<', 1) != -1):
    fixed_data_element = data_element[data_element.find('>', 1)+1:data_element.find('<', 1)]
  else:
    fixed_data_element = data_element[data_element.find('>', 1)+1:]
  return fixed_data_element
```

```python
# We need to scrape the data from OpenCS now
website_url = "https://drafty.cs.brown.edu/csopenrankings/"
html_content_openCS = get_html_content(website_url)
```

```python
# Getting a list of all the universities from the OpenCS website
university_tag = '<td class="column2">'
score_tag = '<td class=column8>'
ending_tag = '</td>'
positions = [match.start() for match in re.finditer(university_tag, html_content_openCS)]
positions_score = [match.start() for match in re.finditer(score_tag, html_content_openCS)]
print(len(positions))
print(len(positions_score))
first_university = html_content_openCS[positions[0]:html_content_openCS.find(ending_tag, positions[0])] # DEBUG
first_score = html_content_openCS[positions_score[0]:html_content_openCS.find(ending_tag, positions_score[0])]
```

```python
print(remove_tags_alt(first_university)) # DEBUG
print(remove_tags_alt(first_score))
```

```python
# Now that we know the simple case works, we want to generate a list of POSITIONS for every page
list_of_universities_openCS = []
for i in range(len(positions)):
  list_of_universities_openCS.append(remove_tags_alt(html_content_openCS[positions[i]:html_content_openCS.find(ending_tag, positions[i])]))
print(list_of_universities_openCS)
```

```python
# List of scores as well
list_of_scores_openCS = []
for i in range(len(positions_score)):
  list_of_scores_openCS.append(remove_tags_alt(html_content_openCS[positions_score[i]:html_content_openCS.find(ending_tag, positions_score[i])]))
print(list_of_scores_openCS)
```

```python
# Creating a dictionary based on opencs data
table = {}
for i in range(len(list_of_universities_openCS)):
  table[list_of_universities_openCS[i]] = int(list_of_scores_openCS[i])

print(table)
```

```python
# Finding if there is a university in the OpenCS entry from the ROI website
# If not, ignore it and check the next university
# Add the score of the ROI onto the score of the OpenCS data

for i in range(len(list_of_universities)):
  if list_of_universities[i] in table:
    table[list_of_universities[i]] = table[list_of_universities[i]] + i
```

```python
print(table)
```

```python
import pandas as pd
```

```python
table_df = {'college': [], 'score': []}
for key, value in table.items():
  table_df['college'].append(key)
  table_df['score'].append(value)
df = pd.DataFrame(table_df)
df
```
